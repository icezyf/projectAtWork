#####2.5 Tokens Analysis#####

```
max_title_df = spark.sql('select Date, Day, Channels, LeadInquiryRate, Title from channel where LeadInquiryRate > 0.6 and Inquiry >= 2')

max_title_df.display()
```

```
max_title_pd = max_title_df.toPandas()
max_title_pd.head()
```

```
title_lst = max_title_pd.loc[:, 'Title'].tolist()
title_lst
```

```
def tokenization(text):
  tokens = []
  for token in text:
    if token == '\r':
      del token
    elif token is None:
      continue
    else:
      tokens.extend(token.split(','))
  return tokens

dup_tokens = tokenization(title_lst)
res=[]
for i in dup_tokens:
  if i not in res:
    res.append(i)

print("These are the keywords in the titles that drove 60% ratio: " + str(res))
```

According from the limit data in Jan, the lead ratio would be and beyond 60% if these the vacancy titles contain the following tokens:
> 'VA', ' DA全职', ' Ai机器人公司', ' SDE实习', 'Google', ' 面经', ' Leetcode', ' 按照', ' 顺序', ' 刷题', ' 比较好', 'Vobile', ' BA实习', '一站式', ' 金融', ' 美国商业银行', ' 金融类', ' BD实习', ' FA实习', ' IA实习', ' Compliance实习'
